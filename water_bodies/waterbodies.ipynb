```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
yell = gpd.read_file('../datos/YELL_tracts/YELL_boundary.shp')
```

```python
yell.plot()
```

```python
yell = yell.to_crs('EPSG:4326')
```

```python
water = gpd.read_file('../datos/USA_Detailed_Water_Bodies/USA_Detailed_Water_Bodies.shp')
```

```python
## filtrar water en yell
water_yell = water[water.intersects(yell.unary_union)]
```
